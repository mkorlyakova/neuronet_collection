```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt 
```

```python
with open('rez_rgb_milani.txt','r') as f:
    s = f.read()
    
s = s.split('\n')
s[0]
```

```python
def parsing_string(s):
    
    step_tr = []
    step_test = [0]
    train_loss  = []
    train_acc = []
    test_loss  = [2.]
    test_acc = [0.5]
    for si in s:
        si = si.split(':')
        #print(si)
        if len(si)>6:
            step = si[5].split('.')
            step_tr += [int(step[0])]
            loss = si[6].split('c')
            train_loss += [float(loss[0][-3])]
            acc = '.'.join(si[7].split('.')[:2])
            train_acc += [float(acc)]

        else:
            step_test += [int(step[0])]
            loss = si[4].split('c')
            test_loss += [float(loss[0][-3])]
            acc = '.'.join(si[5].split('.')[:2])
            test_acc += [float(acc)]
    return step_tr,train_loss,train_acc, step_test, test_loss,test_acc
        
```

```python

with open('rez_rgb_milani.txt','r') as f:
    s = f.read()
    
s = s.split('\n')

step_trr,train_lossr,train_accr, step_testr, test_lossr,test_accr = parsing_string(s)
```

```python
with open('rez_hsv_c.txt','r') as f:
    s = f.read()
    
s = s.split('\n')

step_trhc,train_losshc,train_acchc, step_testhc, test_losshc,test_acchc = parsing_string(s)
```

```python
with open('rez_hsv_milani.txt','r') as f:
    s = f.read()
    
s = s.split('\n')

step_trh,train_lossh,train_acch, step_testh, test_lossh,test_acch = parsing_string(s)
```

```python
plt.figure(figsize=(16,8))
plt.plot(step_testr,test_accr, label = 'rgb val')
plt.plot(step_trr,train_accr, label = 'rgb train')
plt.plot(step_testh,test_acch, label = 'hsv val')
plt.plot(step_trh,train_acch, label = 'hsv train')
plt.plot(step_testhc,test_acchc, label = 'c hsv val')
plt.plot(step_trhc,train_acchc, label = 'c hsv train')
plt.grid()
plt.legend()
plt.xlabel('step')
plt.ylabel('acc')
plt.show()
```
